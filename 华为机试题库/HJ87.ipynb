就是得遍历完所有的情况

```python
def score_length(s):
    l = len(s)
    if l <= 4:
        return 5
    elif l <= 7:
        return 10
    else:
        return 25

def score_letter(s):
    daxie = False
    xiaoxie = False
    
    for c in s:
        if c >= 'a' and c <= 'z':
            xiaoxie = True
        if c >= 'A' and c <= 'Z':
            daxie = True
    
    if daxie == False and xiaoxie == False:
        return 0
    elif daxie == True and xiaoxie == True:
        return 20
    else:
        return 10

def score_number(s):
    n = 0
    for c in s:
        if c >= '0' and c <= '9':
            n += 1
        if n >= 2:
            break
    if n == 0:
        return 0
    elif n == 1:
        return 10
    else:
        return 20


def score_fuhao(s):
    fuhao = ['!','"','#','$','%','&',"'",'(',')','*','+',',','-','.','/',':',';','<','=','>','?','@','[',"]",'^','_','`','{','|','}','~']
    n = 0
    for c in s:
        if c in fuhao or ord(c) in range(0x5B,0x60):
            n += 1
        if n >= 2:
            break
    if n == 0:
        return 0
    elif n == 1:
        return 10
    else:
        return 25

def score_bonus(s):
    flag = 0
    score = 0
    if score_letter(s) != 0 and score_number(s) != 0:
        flag = 1
        score = 2
    if score_letter(s) != 0 and score_number(s) != 0 and score_fuhao(s) != 0:
        flag = 1
        score = 3
    if score_letter(s) == 20 and score_number(s) != 0 and score_fuhao(s) != 0:
        flag = 1
        score = 5
    if flag == 0:
        score = 0
    return score
    
def score(s):
    score = score_length(s) +  score_letter(s) + score_number(s) + score_fuhao(s) + score_bonus(s)
    if score >= 90:
        return 'VERY_SECURE'
    elif score >= 80:
        return 'SECURE'
    elif score >= 70:
        return 'VERY_STRONG'
    elif score >= 60:
        return 'STRONG'
    elif score >= 50:
        return 'AVERAGE'
    elif score >= 25:
        return 'WEAK'
    else:
        return 'VERY_WEAK'

try:
    while True:
        s = input()
        #s = '38$@NoNoNo'
        #s = '123'
        print(score(s))
except:
    pass
```
