```python
import copy

N, K = list(map(int, input().split()))
number = input()

# N = 45
# K = 32

# number = "293440596342887581257444442930778730382520372"

# N = 6
# K = 5

# number = "787585"



list_number = []
for i in range(N):
    list_number.append(int(number[i]))

minsumpay = 10 * len(list_number)

# get minimum pay
for i in range(10):
    list_minus = [abs(n - i) for n in list_number]
    
    list_minus.sort()    
    if sum(list_minus[:K]) < minsumpay:
        minsumpay = sum(list_minus[:K])

print(minsumpay)

    
n_list = []
max_minus_list = []

for i in range(10):
    list_minus = [abs(n - i) for n in list_number]
    
    list_minus.sort()
    if sum(list_minus[:K]) == minsumpay:
        n_list.append(i)
        max_minus_list.append(list_minus[K - 1])
    

numbers_string_list = []
for i in range(len(n_list)):
    
    aim_n = n_list[i]
    max_difference = max_minus_list[i]
    
    new_number_list = copy.copy(list_number)
    sum_pay = 0

    
    for j in range(len(list_number)):
        if abs(new_number_list[j] - aim_n) < max_difference:
            new_number_list[j] = aim_n
            sum_pay += abs(list_number[j] - aim_n)
    
    
    for j in range(len(list_number)):
        if sum_pay == minsumpay:
            break
        elif new_number_list[j] - aim_n == max_difference:
            new_number_list[j] = aim_n
            sum_pay += list_number[j] - aim_n

    for j in range(len(list_number)):
        if sum_pay == minsumpay:
            break
        elif new_number_list[-1 - j] - aim_n == - max_difference:
            new_number_list[-1 - j] = aim_n
            sum_pay += - (list_number[-1 - j] - aim_n)
    
    
    strnum = ""
    for j in range(len(new_number_list)):
        strnum += str(new_number_list[j])
    
    #print(strnum)

    numbers_string_list.append(strnum)

numbers_string_list.sort()

#print(minsumpay)
print(numbers_string_list[0])
```
